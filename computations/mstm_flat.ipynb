```python
from processing import msevn
import os
import pandas as pd

msevn.configure_mpl()
```

```python
root = './matlab/output/flat'

files = os.listdir(root)
```

```python
ext_circ = msevn.mesh_generator.target_plot.project_2d(gap=83*2, edge=16, mult=10, angle=15)
ext_circ = pd.DataFrame(ext_circ, columns=['x', 'y', 'z', 'r'])
```

```python
xz1 = msevn.MstmEngine(os.path.join(root, files[1]))
fig, ax, bbea = xz1.plot_field(
    trim=20,
    figsize=(10, 10),
    target=dict(plot=True, color='magenta', linewidth=0.4, alpha=0.7),
    xtick=3000,
    ytick=2000,
    external_circles=ext_circ,
)
ax.grid(alpha=0.1, color='white')

#fig.savefig(f'../tex/components/img/celes/Es_20nm_15deg_10harm.pdf', bbox_inches='tight')
```

```python
xz0 = msevn.MstmEngine(os.path.join(root, files[0]))
fig, ax, bbea = xz0.plot_field(
    trim=20,
    figsize=(10, 10),
    target=dict(plot=True, color='magenta', linewidth=0.4, alpha=0.7),
    xtick=3000,
    ytick=2000,
    external_circles=ext_circ,
    vmax=0.14,
    factor=0.08,
    bartick=0.02
)
ax.grid(alpha=0.1, color='white')

#fig.savefig(f'../tex/components/img/celes/Es_20nm_15deg_1harm.pdf', bbox_inches='tight')
```
