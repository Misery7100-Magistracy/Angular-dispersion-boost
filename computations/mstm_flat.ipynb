```python
from processing import msevn
import os
import pandas as pd
import numpy as np

msevn.configure_mpl()
```

```python
root = './matlab/output/archive'

files = os.listdir(root)
files
```

```python
ext_circ2 = msevn.mesh_generator.target_plot.project_2d(gap=83*3, edge=3, mult=50, angle=90, onedim=False, thin=True)
ext_circ2 = pd.DataFrame(ext_circ2, columns=['x', 'y', 'z', 'r'])
```

```python
ext_circ = msevn.mesh_generator.target_plot.project_2d(gap=83*3, edge=16, mult=10, angle=90, onedim=True, thin=False)
ext_circ = pd.DataFrame(ext_circ, columns=['x', 'y', 'z', 'r'])
```

```python
xz1 = msevn.MstmEngine(os.path.join(root, files[1]))
fig, ax, bbea = xz1.plot_field(
    trim=10,
    figsize=(10, 10),
    target=dict(plot=True, color='magenta', linewidth=0.4, alpha=0.7),
    xtick=3000,
    ytick=2000,
    external_circles=ext_circ2,
    squared=True,
    vmax=0.6,
    #bartick=0.05,
    custom_func=lambda x: np.log(x + 1),
    custom_label=r'$\ln\left(|\mathbf{E}_{\it{s}}|^2 + 1\right)$',
    angle_compensation=-60,
    fix_limits={'x' : (-6900, 6900), 'y' : (-6900, 6900)},
    custom_x=r'$x^\prime$, $\rm{nm}$',
    custom_y=r'$z^\prime$, $\rm{nm}$'

    )
ax.grid(alpha=0.2, color='white')
#fig.savefig(f'../tex/components/img/celes/3plane_flat_to_compare.pdf', bbox_inches='tight')
```
