```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

x = np.array([5, 9, 15, 20, 30, 40, 50, 60])
y10 = np.array([4.487, 4.761, 4.899, 4.9387, 4.808, 5.072, 5.083, 5.092])
y1 = np.array([7.394, 6.5001, 5.4140, 4.7124, 4.465, 4.365, 4.265, 4.065])


fig, ax = plt.subplots(figsize=(10, 6))
sns.scatterplot(x=x, y=y10 / (2*x), s=130, marker='^', color='purple', label=r'$\lambda=\lambda_{10}$')
sns.scatterplot(x=x, y=y1 / (2*x), s=130, marker='v', color='blue', label=r'$\lambda=\lambda_{L}$')

ax.set_xlabel(r'$a$, $\rm{nm}$', labelpad=15)
ax.set_ylabel(r'$h_{tr}\:\:/\:\:2a$', labelpad=15)
ax.set_ylim(0, 0.8)

#fig.savefig('../tex/components/img/lpic/htr_over_2a_a.pdf', bbox_inches='tight')
```
